```python
import pandas as pd 
import numpy as np 
from covid import Covid 
import matplotlib.pyplot as plt
```

```python
covid = Covid()
covid.get_data()
```

```python
df =  pd.DataFrame(covid.get_data())
```

```python
total_confirmed = df['confirmed'].sum()
total_active = df['active'].sum()
total_recovered = df['recovered'].sum()
total_deaths = df['deaths'].sum()

print(total_confirmed, total_active, total_recovered)
```

```python
labels = ['active', 'recovered', 'deaths']
sizes = [total_active, total_recovered, total_deaths]

fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels = labels)
ax1.axis('equal')
plt.show()
```

```python
df = df.sort_values(by = ['confirmed'], ascending = False)
```

```python
new_df = df[['country', 'confirmed', 'active', 'deaths', 'recovered', 'last_update']]
```

```python
new_df.head(10)
```

```python
graph_top10 = new_df.loc[0:11,['country', 'active']]
graph_top10                    
```

```python
active_cases = graph_top10.loc[:, 'active']
countrys = graph_top10.loc[:, 'country']
```

```python
plt.figure(figsize = (20,10))
plt.barh(countrys, active_cases)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 20)
plt.title('Active cases', fontsize = 25)
plt.show()
```

```python
confirmed_cases = new_df.loc[0:11, 'confirmed']
```

```python
plt.figure(figsize = (20,10))
plt.barh(countrys, confirmed_cases)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 20)
plt.title('Confirmed cases', fontsize = 25)
plt.show()
```

```python
deaths = new_df.loc[0:11, 'deaths']
deaths_per_m = deaths / 1000000
```

```python
plt.figure(figsize = (20,10))
plt.barh(countrys, deaths)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 20)
plt.title('Deaths', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.barh(countrys, deaths_per_m)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 20)
plt.title('Deaths per m', fontsize = 25)
plt.show()
```
